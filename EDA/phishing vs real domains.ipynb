The dataset looks imbalance. Let's use NEARMISS to balance the dataset.

```python
X = df.drop(['phishing'],axis=1)
y = df['phishing']
```

```python
y
```

```python
!pip install imblearn
```

```python
from imblearn.under_sampling import NearMiss
```

```python
sample = NearMiss()
```

```python
X_bal,y_bal = sample.fit_resample(X,y)
```

```python
plot.figure(figsize=(10,5), facecolor='white')
sns.countplot(y_bal)
```
